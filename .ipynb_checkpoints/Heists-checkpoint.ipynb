### The completed function.

This is not ideal since the CDN dosn't appreciate our requests and will block new requests after a while.

```python
def subdl(name, year, filename):
    # GET subdl search results
    url = 'https://api2.subdl.com/search?query='
    query = name.replace(" ", "%20")
    search = url + query

    # attempt to get the search details
    time.sleep(3) # Be kind
    response = requests.get(search)

    if response.status_code != 200:
        print("There was an error from the api", search, response.status_code)
        return False

    parsed_json = json.loads(response.text)

    page_url = ''

    # print(parsed_json['results'])
    for result in parsed_json['results']:
        if result['type'] != 'movie':
            continue
        if result['year'] != year:
            continue
        page_url = '/'.join(['https://subdl.com/subtitle',
                             result['sd_id'],
                             result['slug'],
                             'english'])
        break

    if page_url == '':
        print("    No results found.", search)
        return False

    # attempt to open the subtitle page
    time.sleep(3) # be kind
    response = requests.get(page_url)

    if response.status_code != 200:
        print("    There was an error from the api", page_url, response.status_code)
        return False

    # get all links off of page for subtitles
    soup = BeautifulSoup(response.text, 'html.parser')
    urls = []
    for link in soup.find_all('a'):
        href = link.get('href')
        if href is None:
            continue
        if not href.startswith('https://dl.subdl.com/download/subtitle/'):
            continue
        urls.append(link.get('href'))

    if urls == []:
        print("    No subtitles found on page", page_url)
        return False

    # Define a header for the download
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.2; Win64; x64; rv:16.0.1) Gecko/20121011 Firefox/16.0.1'}

    found_srt = False
    for sub_url in urls:

        time.sleep(3) # be kind
        response = requests.get(sub_url, headers=header)

        if response.status_code != 200:
            print("    Bad response from sub_url", sub_url)
            continue

        # attempt to open the zip file downloaded into memory
        try:
            myzip = ZipFile(BytesIO(response.content))
        except BadZipfile:
            print("    BadZipfile", sub_url)
            continue
        except Exception as err:
            print(f"    Unexpected {err=}, {type(err)=}")
            continue

        for file in myzip.namelist():
            if not file.endswith('.srt'):
                continue

            # and save the file to the filename directory
            with open(filename, 'wb') as binary_file:
                binary_file.write(myzip.open(file).read())
                found_srt = True

        if found_srt:
            break # Found a file, don't continue downloading

    return found_srt
```

## The finished automation

```python
def yifysubtitles(name, year, filename):
    # GET subdl search results
    url = 'https://yifysubtitles.org/search?q='
    query = name.replace(" ", "+")
    search = url + query

    # attempt to get the search details
    time.sleep(3) # Be kind
    response = requests.get(search)

    if response.status_code != 200:
        print("    There was an error downloading the search results.", search, response.status_code)
        return False
    
    soup = BeautifulSoup(response.text, 'html.parser')
    indiatables = soup.find_all('li',{'class':'media media-movie-clickable'})
    
    page_url = ''
    for item in indiatables:
        # print(str(item))
        match = re.search(r"<a href=\"(.*)\" itemprop=\"url\">.*([0-9]{4})<small>year", str(item), re.DOTALL)
        # print(match.group(1),match.group(2))

        if match == None:
            print("Entry doesn't follow the set pattern.", search)
            continue

        if match.group(2) != str(year):
            continue

        page_url = match.group(1)
        break
    
    if page_url == '':
        print("    No results found.", search)
        return False
    
    page_url = 'https://yifysubtitles.org' + page_url
    
    # attempt to open the subtitle page
    time.sleep(3) # be kind
    response = requests.get(page_url)

    if response.status_code != 200:
        print("    There was an error from the page", page_url, response.status_code)
        return False

    # get all links off of page for subtitles
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table')
    
    df_movie = pd.read_html(str(table))
    df_movie = pd.DataFrame(df_movie[0])
    
    urls = []
    for tr in table.findAll("tr"):
        trs = tr.findAll("td")
        for each in trs:
            try:
                url = each.find('a')['href']
                urls.append(url)
            except:
                pass

    df_movie['Link'] = urls[::2]
    
    if urls == []:
        print("    No subtitles found on page", page_url)
        return False
    
    df_ranked = df_movie.loc[df_movie['language'] == 'English'].sort_values('rating', ascending=False)
    
    # Define a header for the download
    # header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101 Firefox/50.0'}
    
    found_srt = False
    for r_index, r_row in df_ranked.iterrows():
        sub_url = 'https://yifysubtitles.org' + r_row['Link'] + '.zip'
        sub_url = sub_url.replace('/subtitles/','/subtitle/')
        
        time.sleep(3) # be kind
        response = requests.get(sub_url)

        if response.status_code != 200:
            print("    Bad response from sub_url", sub_url)
            continue

        # attempt to open the zip file downloaded into memory
        try:
            myzip = ZipFile(BytesIO(response.content))
        except BadZipfile:
            print("    BadZipfile", sub_url)
            continue
        except Exception as err:
            print(f"    Unexpected {err=}, {type(err)=}")
            continue

        for file in myzip.namelist():
            if not file.endswith('.srt'):
                continue

            # and save the file to the filename directory
            with open(filename, 'wb') as binary_file:
                binary_file.write(myzip.open(file).read())
                found_srt = True

        if found_srt:
            break # Found a file, don't continue downloading

    return found_srt
```

## Applied to our Heist movies list

The great subtitle heist begins!

```python
# Fill subs column

for index, row in df.iterrows():
    
    name = row['Film']
    year = row['Year']
    
    # prevents running everything again
    if not row['Subs'] == '':
        continue
    
    filename = 'subs/' + name + ' (' + str(year) + ')' + '.srt'
    
    if exists(filename):
        print(name, "subtitles already exist.")
        
        with open(filename, 'r', errors="ignore") as f:
            data = f.read()
            df.loc[index,'Subs'] = data

    else: # Try and find subtitles online
        
        print("Searching for subtitles for:", name, year)

        # subdl_found = subdl(name, year, filename)
        # if not subdl_found:
        #     print("    No subtitles found on subdl.")
        #     continue
        
        yify_found = yifysubtitles(name, year, filename)
        if not yify_found:
            print("    No subtitles found on yify.")
            continue
        
        print("    ", name, "subtitles downloaded.")
        with open(filename, 'r', errors="ignore") as f:
            data = f.read()
            df.loc[index,'Subs'] = data           
            
        
```

```python
def yifysubtitles_lengths(name, year):
    # GET subdl search results
    url = 'https://yifysubtitles.org/search?q='
    query = name.replace(" ", "+")
    search = url + query

    # attempt to get the search details
    response = requests.get(search)

    if response.status_code != 200:
        print("    There was an error downloading the search results.", search, response.status_code)
        return False
    
    soup = BeautifulSoup(response.text, 'html.parser')
    indiatables = soup.find_all('li',{'class':'media media-movie-clickable'})
    
    length = 0.0
    for item in indiatables:
        # print(str(item))
        match = re.search(r"<a href=\"(.*)\" itemprop=\"url\">.*([0-9]{4})<small>year.*\"movinfo-section\">(\d*)<small>min", str(item), re.DOTALL)
        # print(match.group(1),match.group(2))

        if match == None:
            # print("Entry doesn't follow the set pattern.", search)
            continue

        if match.group(2) != str(year):
            continue

        length = match.group(3)
        break
    
    return length
```

```python
for index, row in df.iterrows():
    
    name = row['Film']
    year = row['Year']
    
    df.loc[index, 'Minutes'] = yifysubtitles_lengths(name, year)
```

```python
df['Minutes'] = df['Minutes'].astype('float')
df.loc[df['Minutes'] == 0.0]
```

```python
df['Freq'] = df['Looks']/df['Minutes'].replace({ 0 : np.nan })
```
